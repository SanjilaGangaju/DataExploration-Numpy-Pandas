# Import Libraries

Create pie charts and bar graphs from data that you collect from members of your Code Club

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
y=np.array([35,25,25,12,3])
mylabels=["Cat","Dog","Hampster","Fish","Snake"]
plt.pie(y,labels=mylabels)
plt.show()
```

# Edit Graph:Explode

```python
y=np.array([35,25,25,15,2])
mylabels=["Cat","Dog","Hampster","Fish","Snake"]
myexplode=[0.1,0,0,0,0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

# Create Bar Charts

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["Cat", "Dog", "Hampster", "Fish", "Snake"])
y=np.array([3, 8, 1, 10, 4])

plt.bar(x,y)
plt.show()
```

# Edit Graph :Colour

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["Cat", "Dog", "Hampster", "Fish", "Snake"])
y=np.array([3, 8, 1, 10, 4])
plt.bar(x,y,color="pink")
plt.show()
```

# Edit Graph Width

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["Elephant","Deer","Tiger","Lion"])
y=np.array([3,6,23,20])

plt.bar(x,y,color= "blue",width = 0.4)
plt.show()
```

# Edit Graph Rotate

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["Elephant","Deer","Tiger","Lion"])
y=np.array([3,6,23,20])

plt.barh(x,y)
plt.show()
```
